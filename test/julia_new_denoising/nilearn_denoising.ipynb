```python
from nilearn.image import high_variance_confounds
from nilearn.input_data import NiftiMasker
import numpy as np
from scipy.special import legendre
import matplotlib.pyplot as plt
```

```python
subjects=['GAET']
sessions=['rest1_1']
```

```python
for subject in subjects:
    for session in sessions:

        base_dir='/scr/ilz3/myelinconnect/resting/preprocessed/'+subject+'/'+session+'/'
        raw_file=base_dir+'realignment/corr_GAET_sub005_rest1_1_roi.nii.gz'
        moco_file=base_dir+'realignment/GAET_sub005_rest1_1_roi.nii.gz'
        brain_mask=base_dir+'mask/GAET121203_S4_mp2rage_0.5iso_T1_Images_mask_fixed_trans.nii.gz'
        wm_mask=base_dir+'mask/GAET121203_S5_mp2rage_0.5iso_UNI_Images_merged_pve_0_trans_maths.nii.gz'
        csf_mask=base_dir+'mask/GAET121203_S5_mp2rage_0.5iso_UNI_Images_merged_pve_2_trans_maths.nii.gz'
        motion_file_12=base_dir+'confounds/motion_regressor_der1_ord1.txt'
        motion_file_24=base_dir+'confounds/motion_regressor_der1_ord2.txt'
        artefact_file=base_dir+'confounds/art.corr_GAET_sub005_rest1_1_roi_outliers.txt'
        out_file=base_dir+'rest_denoised_%s.nii.gz'
        confound_file=base_dir+'confounds/all_confounds_%s.txt'
        compcor_file=base_dir+'confounds/compcor_regressor.txt'
```

```python
confound_len = nb.load(moco_file).get_data().shape[3]
```

```python
# reload niftis to round affines so that nilearn doesn't complain
wm_nii=nb.Nifti1Image(nb.load(wm_mask).get_data(), around(nb.load(wm_mask).get_affine(), 2), nb.load(wm_mask).get_header())
csf_nii=nb.Nifti1Image(nb.load(csf_mask).get_data(),around(nb.load(csf_mask).get_affine(), 2), nb.load(csf_mask).get_header())
moco_nii=nb.Nifti1Image(nb.load(moco_file).get_data(),around(nb.load(moco_file).get_affine(), 2), nb.load(moco_file).get_header())
```

```python
# extract high variance confounds from unprocessed data
highvar_regressor=high_variance_confounds(raw_file, detrend=True)
```

```python
# extract high variance confounds in wm/csf masks from motion corrected data
wm_regressor=high_variance_confounds(moco_nii, mask_img=wm_nii, detrend=True)
csf_regressor=high_variance_confounds(moco_nii, mask_img=csf_nii, detrend=True)
```

```python
# load compcor regressors
compcor_regressor=np.genfromtxt(compcor_file)
```

```python
# load motion regressors
motion_regressor_12=np.genfromtxt(motion_file_12)
motion_regressor_24=np.genfromtxt(motion_file_24)
```

```python
# create outlier regressors
outlier_regressor = np.empty((confound_len,1))
try:
    outlier_val = np.genfromtxt(artefact_file)
except IOError:
    outlier_val = np.empty((0))
for index in np.atleast_1d(outlier_val):
    outlier_vector = np.zeros((confound_len, 1))
    outlier_vector[index] = 1
    outlier_regressor = np.hstack((outlier_regressor, outlier_vector))

outlier_regressor = outlier_regressor[:,1::]
```

```python
# create regressors for linear and quadratic trends (and constant?)
trend_regressor = np.ones((confound_len, 1))
for i in range(2):
    trend_regressor = np.hstack((trend_regressor, legendre(i + 1)(np.linspace(-1, 1, confound_len))[:, None]))
```

### compare nilearn and regressors detrending

```python
plt.figure(figsize=(5,5))
plt.title('trend regressors')
plt.plot(trend_regressor)
plt.ylim((-1,1.5))
```

```python
denoiser_detrend=NiftiMasker(mask_img=brain_mask, standardize=True, detrend=True)
denoiser_nodetrend=NiftiMasker(mask_img=brain_mask, standardize=True, detrend=False)
```

```python
nilearn_detrend=denoiser_detrend.fit_transform(moco_file, confounds=None)
regressor_detrend=denoiser_nodetrend.fit_transform(moco_file, confounds=trend_regressor)
```

```python
vxl = np.random.randint(0,(nilearn_detrend.shape[1]+1))
plt.figure(figsize=(20,10))
plt.plot(nilearn_detrend[:,vxl], label='nilearn detrending', color='b')
plt.plot(regressor_detrend[:,vxl], label='regressor detrending', color='r')
plt.legend()
plt.title('compare detrending')
```
